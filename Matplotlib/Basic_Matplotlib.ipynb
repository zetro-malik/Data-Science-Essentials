```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max') # defining labels for legend
plt.plot(days,min_t,label='min') 
plt.plot(days,avg_t,label='avg') 

plt.legend(loc='lower left', shadow=True, fontsize='large') # Legend, popUp telling bout each Line, by default 'best'

```

#### Applying grid

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max', marker='D',markersize=12) # defining labels for legend
plt.plot(days,min_t,label='min',marker='D',markersize=12) 
plt.plot(days,avg_t,label='avg',marker='D',markersize=12) 

plt.legend(loc='lower left', shadow=True, fontsize='large') # Legend, popUp telling bout each Line, by default 'best'

plt.grid()
```

#### <font color="Maroon"><h4 align="Left">Bar Chart

```python
import numpy as np 
```

### Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
# plt.bar does not take string List
xpos = np.arange(len(company))
plt.bar(xpos,revenue)
```

```python
# To replace dummy ypos values with actual Strings name
plt.xticks(xpos,company) # to replace ypos with company
plt.ylabel('Revenue(bln $)')
plt.xlabel('Company')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Rev')

plt.legend()
```

### Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python

plt.xticks(xpos,company)
plt.ylabel('Revenue(bln $)')
plt.xlabel('Company')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='profit')

plt.legend()
```

####  what if I want side by side Bars ?


```python
# minus revenue, profit by a little margin from Xpos and min the width
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln $)')
plt.xlabel('Company')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,label='Revenue' ,width=0.4)
plt.bar(xpos+0.2,profit,label='profit',width=0.4)

plt.legend()
```

#### For horizontal bars

```python
# minus revenue, profit by a little margin from Xpos and min the width
plt.yticks(xpos,company)
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,label='Revenue', height=0.4)
plt.barh(xpos+0.2,profit,label='profit', height=0.4)

plt.legend()
```

## <font color="Maroon"><h4 align="center">Histograms</h4></font>
#### In histograms, x axis contains a variable and y axis will be a frequency of that variable
#### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=3, rwidth=0.95) # by default number of bins is set to 10, bins is number of ranges
```

```python
# specify bins ranges and different parameters for Hist()
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95 ,color='g', histtype='bar', align='mid',orientation='vertical') 
```

#### Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

## <font color="Maroon"><h4 align="center">Pie Charts</h4></font>

### Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
```

#### Draw pie chart as perfect circle

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

#### Show percentages for every pie. Specify radius to increase chart size

```python
# 0.0 for no decimal points
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5)
plt.show()
```

#### Explode

```python
# 0.0 for no decimal points
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5, explode=[0,0.15,0,0,0])
plt.show()
```

#### counterclock and angle properties


```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.savefig("figure.png", dpi=600)
plt.show()
```
