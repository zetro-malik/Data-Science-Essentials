```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,52,51]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color="green",linewidth=5,linestyle='dotted')
```

```python
plt.plot(x,y,'g--')
```

```python
# color, marker on (x,y), linestyle(dotted,dashed..), markersize, alpha 
plt.plot(x,y,color="red",marker="D",linestyle="",markersize=12,alpha=0.5)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max') # defining labels for legend
plt.plot(days,min_t,label='min') 
plt.plot(days,avg_t,label='avg') 

plt.legend(loc='lower left', shadow=True, fontsize='large') # Legend, popUp telling bout each Line, by default 'best'

```

#### Applying grid

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max', marker='D',markersize=12) # defining labels for legend
plt.plot(days,min_t,label='min',marker='D',markersize=12) 
plt.plot(days,avg_t,label='avg',marker='D',markersize=12) 

plt.legend(loc='lower left', shadow=True, fontsize='large') # Legend, popUp telling bout each Line, by default 'best'

plt.grid()
```

### Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
# plt.bar does not take string List
xpos = np.arange(len(company))
plt.bar(xpos,revenue)
```

```python
# To replace dummy ypos values with actual Strings name
plt.xticks(xpos,company) # to replace ypos with company
plt.ylabel('Revenue(bln $)')
plt.xlabel('Company')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Rev')

plt.legend()
```
